## **Zadanie nr 5 - wizualizacja danych**

> Korzystając z biblioteki MatPlotLib mamy możliwość wizualizacji danych. Biblioteka ta pozwala na tworzenie różnego rodzaju wykresów (liniowych, słupkowych itd.). Bez problemu można na jednym wykresie przedstawić dwie lub więcej serie danych.

> [Przykłady wykorzystania wykresów](https://matplotlib.org/3.1.1/gallery/)

Na poniższym wykresie kołowym przedstawiono udział liczby aktualnie zakażonych, ozdrowieńców i zmarłych w całkowitej liczbie wykrytych przypadków w Polsce.

```
import matplotlib
import matplotlib.pyplot as plt

data = df.copy()
#calkowita liczba wykrytych przypadków
pl_confirmed = data.loc[data['country/region']=='Poland', 'confirmed'].iloc[0]
#aktualnie chorzy
pl_active = data.loc[data['country/region']=='Poland', 'active'].iloc[0]
#liczba zgonow
pl_deaths = data.loc[data['country/region']=='Poland', 'deaths'].iloc[0]
#liczba wyzdrowialych
pl_confirmed = data.loc[data['country/region']=='Poland', 'recovered'].iloc[0]

#procentowy udział
pl_active_percent = (pl_active/pl_confirmed)*100
pl_deaths_percent = (pl_deaths/pl_confirmed)*100
pl_confirmeds_percent = (pl_confirmed/pl_confirmed)*100

labels = ['Liczba aktywnych przypadków', 'Liczba zmarłych', 'Liczba wyzdrowiałych']

sizes = [pl_active_percent, pl_deaths_percent, pl_confirmeds_percent]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.2f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

**5a**. Stwórz wykres słupkowy przedstawiający udział liczby aktywnych przypadków w 10 krajach, w których ta liczba jest największa. Dodaj legendę, tytuł wykresu i tytuł osi x. Zapisz wygenerowany wykres do pliku o nazwie **"covid19_top10.jpg"**.

```
# tutaj umieść kod
labels = df.loc[df['active'].nlargest(10).index,"country/region"]
values = df["active"].nlargest(10)
plt.ylabel('Most active cases')
plt.xlabel('Countries')
plt.bar(labels,values, color='violet')
plt.legend(['Active cases'])
plt.title("Top 10 countries with most active cases")
plt.show()
```

Testując poniższy przykład można sprawdzić, jak wyglądają dwa rodzaje przypisania tej samej tablicy.

```
kopia = jednowymiarowa.copy() #tworzymy kopię tablicy
referencja = jednowymiarowa #tworzymy referencję do tablicy

# zmieniamy wartość w kopii tablicy jednowymiarowej
kopia[1] = 100

# wyświetlamy każdą z tablic
print(jednowymiarowa)
print(referencja)
print(kopia)
```

```
# zmieniamy wartość w referencji tablicy jednowymiarowej
referencja[1] = 200

# wyświetlamy każdą z tablic
print(jednowymiarowa)
print(referencja)
print(kopia)
```

```
#czy kopia jest tym samym co  tablica jednowymiarowa
kopia is jednowymiarowa

#czy referencja jest tym samym co  tablica jednowymiarowa
referencja is jednowymiarowa
```

**6f**. Biblioteka numpy pozwala na wykonywanie działań na tablicach, których kształty są różne (broadcasting).

> Mechanizm broadcastingu może być zastosowany do dwóch tablic, jeśli ich wymiary są kompatybilne - tj. ten sam wymiar w dwóch tablicach jest równy lub wynosi 1.

Na poniższym przykładzie zademonstrowany został ten mechanizm.




```
# tworzymy dwie tablice o różnych wymiarach
br_1 = np.zeros(4)
br_2 = np.ones(7)

# wykonanie dodawania (br_1 + br_2) na tak stworzonych tablicach spowoduje wyświetlenie:
#"ValueError: operands could not be broadcast together with shapes (4,) (7,)"

#aby ta operacja mogła zostać wykonana, zmieńmy tablicę br_2 tak, aby miała 7 wierszy i jedną kolumnę
br_2.resize(7,1)

#teraz wymiary obu tablic są kompatybilne - możemy "rozciągnąć" tablicę br_1 aby miała 7 wierszy, a br_2 tak,
#aby miała 4 kolumny
br_1 + br_2
```
